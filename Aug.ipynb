```python
dpr_a=dpr9[['Relationship_Final_name','Month','Premium']]
dpr_a.head(1)
```

```python
dpr_a['Relationship_Final_name']=dpr_a['Relationship_Final_name'].astype(str)
dpr_a['Relationship_Final_name']=dpr_a['Relationship_Final_name'].str.upper()
dpr_a['Month']=dpr_a['Month'].astype(str)
dpr_a['Premium']=dpr_a['Premium'].astype(float)
```

```python
dpr_a=dpr_a.groupby(['Relationship_Final_name','Month']).sum()
dpr_a=dpr_a.reset_index(level=['Relationship_Final_name','Month'])
print(dpr_a.shape)
dpr_a.head(1)
```

```python
dpr_a['Average'] = dpr_a.groupby(['Relationship_Final_name','Month'])['Premium'].transform('mean')
dpr_a['Rank'] = dpr_a['Average'].rank(method='dense', ascending=False)
dpr_a['Rank']=dpr_a['Rank'].astype(int)
dpr_a=dpr_a.sort_values(by='Rank', ascending=True)
dpr_a['Premium']=dpr_a['Premium'].round(5)
dpr_a['Average']=dpr_a['Average'].round(5)
dpr_a.head(1)
```

```python
dpr_a['Percentage']=(dpr_a['Average']/(dpr_a.Premium.sum()))*100
dpr_a['Percentage']=dpr_a['Percentage'].astype(float)
dpr_a['Percentage']=dpr_a['Percentage'].round(5)
dpr_a.head(1)
```

```python
dpr_a=dpr_a.reset_index()
dpr_a=dpr_a.drop(columns=['index'])
dpr_a.head(1)
```

```python
dpr_a['sum'] = dpr_a['Percentage'].shift(0).cumsum().fillna(dpr_a['Percentage'].iat[0])
dpr_a['sum_1'] = dpr_a['sum']
dpr_a['sum_1'] = dpr_a['sum_1'].astype(int)
dpr_a.head(3)
```

```python
dpr_a['bucket']=0
dpr_a.loc[dpr_a.sum_1 <= 25,'bucket'] = '0-25'
dpr_a.loc[(dpr_a.sum_1 > 25)&(dpr_a.sum_1 <= 50),'bucket'] = '26-50'
dpr_a.loc[(dpr_a.sum_1 > 50)&(dpr_a.sum_1 <= 75),'bucket'] = '51-75'
dpr_a.loc[(dpr_a.sum_1 > 75),'bucket'] = '76-100'
dpr_a.head(3)
```

```python
dpr_a=dpr_a.rename(columns={'Premium':'Premium_groupby'})
print(dpr_a.shape)
dpr_a.head(1)
```
